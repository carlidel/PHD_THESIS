```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# use latex in matplotlib
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
x = np.linspace(-2, 2, 1000)
```

```python
x = np.linspace(0, 1, 1000)

plt.figure(figsize=(4, 4))

color_list = ['red', 'blue', 'green', 'black', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

for i, c in list(zip(range(2, 5), color_list))[::-1]:
    for j in (range(1, i)):
        nx = j
        ny = i - j
        for q in range(0, i+1):
            plt.plot(x, q/ny - nx / ny * x, c=c)
            plt.plot(x, q/ny - nx / -ny * x, c=c)
            plt.plot(x, q/-ny - nx / -ny * x, c=c)
            plt.plot(x, q/-ny - nx / -ny * x, c=c)

            plt.axvline(j/i, c=c, linestyle='dotted')
            plt.axhline(j/i, c=c, linestyle='dotted')

plt.xlim(0,1)
plt.ylim(0,1)

plt.xlabel("$(\\omega_x / 2\\pi) \\;\\mathrm{mod}\\; 1$")
plt.ylabel("$(\\omega_y / 2\\pi) \\;\\mathrm{mod}\\; 1$")

plt.plot([], [], c=color_list[0], label="Resonance order 2")
plt.plot([], [], c=color_list[1], label="Resonance order 3")
plt.plot([], [], c=color_list[2], label="Resonance order 4")

# render these two points on top of the lines
plt.scatter([0.28], [0.31], marker="*", c="k", s=100, zorder=10, label="LHC injection tune: $(\\omega_x, \\omega_y) = (0.28, 0.31)$")
plt.scatter([0.31], [0.32], marker=".", c="k", s=150, zorder=10, label="LHC collision tune: $(\\omega_x, \\omega_y) = (0.31, 0.32)$")

plt.tight_layout()
# add legend outside of plot on the right
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)

plt.savefig("2_accelerator_physics_fundamentals/figs/tune_space.pdf", bbox_inches='tight', pad_inches=0.1, dpi=600)
```
